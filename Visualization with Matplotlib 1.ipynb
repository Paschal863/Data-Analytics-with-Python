```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
#plotting a line graph to show Gas Price of different countries over time

plt.figure(figsize=(8,5))
plt.title('Gas Prices Over Time (in USD)', fontdict= {'fontweight':'bold','fontsize':17} )

plt.plot(gas.Year, gas.USA, 'r.-', label = 'USA')
plt.plot(gas.Year, gas.Canada, 'b.-', label = 'Canada' )
plt.plot(gas.Year, gas.France, 'g.-', label = 'France' )
plt.plot(gas.Year, gas.Mexico, 'y.-', label = 'Mexico' )

plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.legend()

plt.savefig('Gas_price_figure.png', dpi = 300) 
#to increase the pixel quality

plt.show()
```
