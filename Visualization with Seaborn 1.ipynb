```python
import seaborn as sns
```

```python
#getting a dataset

sns.get_dataset_names()
```

```python
tips = sns.load_dataset ('tips')
iris = sns.load_dataset ('iris')
titanic = sns.load_dataset ('titanic')
planets = sns.load_dataset ('planets')
```

```python
tips
```

```python
#plotting using scatterplot

sns.scatterplot(x='tip', y='total_bill',data=tips, hue = 'day', size='size')
```

```python
#histogram

sns.histplot(tips['tip'], kde = True, bins=15)
```

```python
#barchart

sns.barplot(x='sex', y='tip', data=tips)
```

```python
#boxplot
sns.boxplot(x='day', y='tip',data=tips, hue = 'sex', palette='YlGnBu')
```

```python
#stripplot
sns.stripplot(x='day', y='tip', data=tips, hue='sex',dodge=True)#dodge is used to separate the gender
```
